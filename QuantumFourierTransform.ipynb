```python
from qiskit import *
from math import pi
```

```python
def myQFT(nqubits):
    qc = QuantumCircuit(nqubits)
    
    for qubit in range(nqubits):
        qc.h(qubit)
        for qu in range(qubit+1,nqubits):
            qc.cu1(pi/2**(qu - qubit),qu,qubit)
    return qc
   
```

```python
display(myQFT(4).draw('mpl'))
```

```python
display(circuit.library.QFT(4).draw('mpl'))
```
